# Import necessary libraries


```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors

```

# Sample user preferences (can be replaced with actual data)

```python
user_preferences = {
    'calories': 2000,
    'protein': 100,
    'carbs': 300,
    'fat': 50,
    'allergies': ['peanuts', 'shellfish'],
    'dietary_restrictions': ['vegetarian']
}
```

# Sample recipe database (can be replaced with actual data)

```python
recipes = {
    'recipe_id': [1, 2, 3, 4, 5],
    'calories': [300, 400, 250, 350, 320],
    'protein': [20, 30, 15, 25, 22],
    'carbs': [40, 50, 30, 45, 35],
    'fat': [10, 15, 8, 12, 11],
    'ingredients': [['chicken', 'rice', 'vegetables'], ['tofu', 'quinoa', 'vegetables'], ['pasta', 'tomato sauce'], ['salmon', 'brown rice', 'broccoli'], ['beans', 'rice', 'avocado']],
    'tags': [['gluten-free'], ['vegetarian'], [], ['pescatarian'], ['vegetarian', 'vegan']]
}
```

# Convert user preferences and recipe database to DataFrames


```python
user_df = pd.DataFrame([user_preferences])
recipe_df = pd.DataFrame(recipes)
```

# Feature scaling on recipe data

```python
scaler = StandardScaler()
recipe_features = scaler.fit_transform(recipe_df[['calories', 'protein', 'carbs', 'fat']])
```

# Find nearest recipes based on user preferences


```python
nn = NearestNeighbors(n_neighbors=3)
nn.fit(recipe_features)
```

# Transform user preferences to scaled values


```python
scaled_preferences = scaler.transform(user_df[['calories', 'protein', 'carbs', 'fat']])
```

# Find nearest recipes


```python
nearest_recipes_idx = nn.kneighbors(scaled_preferences)[1][0]
```

# Get recommended recipes


```python
recommended_recipes = recipe_df.iloc[nearest_recipes_idx]
```

Output

```python
print("Recommended Recipes:")
print(recommended_recipes[['recipe_id', 'calories', 'protein', 'carbs', 'fat', 'ingredients', 'tags']])
```
